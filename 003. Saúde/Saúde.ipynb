```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Health.csv')
```

```python
df = df.reset_index(drop=True)
df
```

```python
densidade_demografica = df.sort_values(by = 'LifeExpectancy', ascending=False).reset_index(drop=True)
densidade_demografica_1 = densidade_demografica[['Country', 'LifeExpectancy', 'Region']].query('(Region == "East Asia & Pacific")')
sns.kdeplot(data=densidade_demografica_1, x='LifeExpectancy', shade=True)

plt.title('DistribuiÃ§Ã£o da Expectativa de Vida na RegiÃ£o East Asia & Pacific', fontsize=16, fontweight='bold')
mean_life_expectancy = densidade_demografica_1['LifeExpectancy'].mean()
plt.axvline(mean_life_expectancy, color='blue', linestyle='dashed', linewidth=1)
plt.text(mean_life_expectancy + 0.5, 0.01, 'MÃ©dia', color='blue', fontsize=12)
plt.tight_layout()


plt.show()

#grafico de densidade baseada nos paÃ­ses com maior expectativa de vida da East Asia & Pacific
```
