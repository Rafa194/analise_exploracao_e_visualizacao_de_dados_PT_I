```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('indices_pobreza_consolidado.csv')
```

```python
df
```

```python

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='porcentagem_pobreza', bins=13, kde=True, color='purple')
plt.xlabel('Porcentagem de Pobreza')
plt.ylabel('FrequÃªncia')
plt.title('DistribuiÃ§Ã£o da Porcentagem de Pobreza em 10 anos')
plt.show()
```

```python

df['referencia'] = pd.to_datetime(df['referencia'], format='%m/%Y')
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x=df['referencia'].dt.year, y='porcentagem_pobreza', marker='o', color='skyblue')
plt.xlabel('Ano')
plt.ylabel('Porcentagem de Pobreza')
plt.title('Porcentagem de Pobreza ao Longo dos Anos')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
```

```python


df['referencia'] = pd.to_datetime(df['referencia'], format='%m/%Y')
plt.figure(figsize=(10, 6))
sns.lineplot(data=df, x=df['referencia'].dt.year, y='populacao_estimada', marker='*', color='red')
plt.xlabel('Ano')
plt.ylabel('PopulaÃ§Ã£o')
plt.title('EvoluÃ§Ã£o da quantidade populacional ao longo dos anos')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
